```python
def fileGenerator(file1, file2, fileSavePath):
    file1 = pd.read_parquet(file1)
    file2 = pd.read_parquet(file2)
    file1_2_mean_earnings = (file1['earnings'] + file2['earnings']) / 2
    copy_file = file1.copy()
    copy_file['earnings'] = file1_2_mean_earnings
    copy_file.to_parquet(fileSavePath, index=False)
```

```python
fileGenerator("earnings_date=2022-02-11/employee_earnings.parquet", "earnings_date=2022-02-12/employee_earnings.parquet", "earnings_date=2022-02-16/employee_earnings.parquet")
```
