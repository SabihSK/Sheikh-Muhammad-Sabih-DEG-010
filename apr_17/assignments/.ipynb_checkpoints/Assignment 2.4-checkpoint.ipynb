```python
import pandas as pd
import plotly.express as px
```

```python
data_set = pd.read_csv("data.csv")
data_set
```

```python
features_list = [
    "area_mean",
    "texture_worst",
    "area_worst",
    "compactness_worst",
    "symmetry_worst",
]
```

```python
show_data = px.scatter_matrix(
    data_frame=data_set,
    dimensions=features_list,
    height=1000,
    color="diagnosis",
    color_discrete_sequence=["#43cea2", "#185a9d"],
)
show_data.show()
```
